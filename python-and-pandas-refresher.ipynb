```python
# if the column names you want to match are named differently, use the left_on and right_on parameters
```

```python
raw_data = {
        'id': ['1', '2', '3', '4', '5', '6', '7'],
        'first_name': ['Jayson', 'Lebron', 'Kevin', 'Stephen', 'Nikola', 'Chris', 'Joel'], 
        'last_name': ['Tatum', 'James', 'Durant', 'Curry', 'Jokic', 'Paul', 'Embiid']}
```

```python
df_1 = pd.DataFrame(raw_data, columns = ['id', 'first_name', 'last_name'])
df_1
```

```python
raw_data = {
        'player_id': ['1', '2', '3', '4', '5'],
        'career_ppg': [24.2, 25.7, 26.2, 25.9, 24.8]}
```

```python
df_2 = pd.DataFrame(raw_data, columns = ['player_id','career_ppg'])
df_2
```

```python
df_3 = pd.merge(df_1, df_2, how = 'right', left_on = 'id', right_on = 'player_id')
df_3
```

```python
# drop the duplicate id column
df_4 = pd.merge(df_1, df_2, how = 'right', left_on = 'id', right_on = 'player_id').drop('id', axis = 1)
df_4
```

```python
player_stats = {'player_name': ['Tatum', 'Durant', 'Butler', 'Davis', 'Jokic'], 
        'season_id': ['2022-23', '2022-23', '2022-23', '2022-23', '2022-23'], 
        'total_points': [2423, 2390, 495, 539, 1928],
        'games_played': [78, 65, 51, 32, 69]}
df = pd.DataFrame(player_stats)
```

```python
df
```

```python
def calc_ppg(row):
    gp = row[3]
    points = row[2]
    return points / gp
```

```python
df['ppg'] = round(df.apply(calc_ppg, axis=1), 1)
df
```
